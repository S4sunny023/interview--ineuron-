```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
```

```python
# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Normalize the pixel values of the images
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# Add a channels dimension to the images
train_images = train_images[..., tf.newaxis]
test_images = test_images[..., tf.newaxis]
```

```python
# Create a small CNN model with less than 10,000 trainable parameters
model = models.Sequential()
model.add(layers.Conv2D(24, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(12, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(12, (1, 1), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(22, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='SGD',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, batch_size=32, epochs=100, validation_split=0.2)
```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f'Test accuracy: {test_acc * 100:.2f}%')
```
